```python
transactions = pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/transactions.csv')
transactions.head(2)
```

12. Joining Users to Transactions using LEFT

```python
leftTrans = pd.merge(transactions, users, how='left')
leftTrans
```

13. Transactions where userid not in Users

```python
transactions[leftTrans['User'].isnull()]
```

14. Users to transactions matching via UserID using Inner

```python
matchUserTrans = pd.merge(transactions, users, how='inner')
matchUserTrans
```

15. Matching and unmatching rows

```python
matchUnmatch = pd.merge(transactions, users, how='outer')
matchUnmatch
```
